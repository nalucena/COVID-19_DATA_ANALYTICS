```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_2020 = pd.read_csv('cases-brazil-cities-time_2020.csv')
base_2020
```

```python
base_2020.drop(columns=['ibgeID', 'deaths_per_100k_inhabitants', 'totalCases_per_100k_inhabitants', 'deaths_by_totalCases'])
```

```python
filtered = (base_2020['state'] != 'TOTAL') & (base_2020['deaths'] != 0)
base_2020_filtered = base_2020[filtered]
base_2020_filtered.drop(columns=['ibgeID', 'city', 'deaths_per_100k_inhabitants', 'totalCases_per_100k_inhabitants', 'deaths_by_totalCases'])
```

```python
base_2020_filtered['deaths'].max()
```

```python
df_base_2020 = base_2020_filtered[['date','newDeaths','newCases']].groupby('date').sum()
df_base_2020
```
